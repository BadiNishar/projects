```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
df=pd.read_csv('weight-height.csv')
```

```python

# df.drop(columns='Gender',inplace=True)
df.head()
```

```python
# scatter plot

plt.scatter(df['Weight'], df['Height'])
plt.xlabel('weight')
plt.ylabel('height')
```

```python
# Finding correlation

df.corr()
```

```python
# seaborn for visualization

import seaborn as sns
sns.pairplot(df)
```

```python
# Independent and dependent features

x=df[['Weight']] #Independent variable should be data frame or two dimensional array

y=df['Height'] # dependent feature can be in series or one dimension
```

```python
# train test split

from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train.shape
```

```python
# Standardization

from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
x_train=scaler.fit_transform(x_train)
```

```python
x_train
```

```python
x_test=scaler.transform(x_test)
```

```python
x_test
```

```python
# Apply Linear Regression

from sklearn.linear_model import LinearRegression
```

```python
regression=LinearRegression()
```

```python
regression.fit(x_train,y_train)
```

```python
# coefficient

print(f'coefficient or slope : {regression.coef_}')
print(f'intercept : {regression.intercept_}')
```

```python
# Plot best fit line on training data

plt.scatter(x_train, y_train)
plt.plot(x_train, regression.predict(x_train))
```

```python
# prediction for test data

y_pred = regression.predict(x_test)
```

```python
# performance metrics
from sklearn.metrics import mean_absolute_error,mean_squared_error
```

```python
mse=mean_squared_error(y_test,y_pred)
mae=mean_absolute_error(y_test,y_pred)
rmse=np.sqrt(mse)
print(mse)
print(mae)
print(rmse)
```

```python
from sklearn.metrics import r2_score
```

```python
R_score=r2_score(y_test,y_pred)
R_score
```

```python
# prediction for new value

x1=regression.predict(scaler.transform([[172.31]]))
x2=regression.predict(scaler.transform([[158.65]]))
x3=regression.predict(scaler.transform([[145.85]]))
x4=regression.predict(scaler.transform([[150]]))
x5=regression.predict(scaler.transform([[165]]))
```
